## Variables and Data Types

### What Are Variables Really?

Think of variables as **labeled boxes** where you can store information. Python has several basic types.

### Numbers, Strings, Boolens

```python
# Numbers
age = 25
height = 5.9
temperature = -10

print(f"Age: {age} (type: {type(age).__name__})")
print(f"Height: {height} (type: {type(height).__name__})")
print(f"Temperature: {temperature}")
```

```python
# Strings
name = "Alice"
message = "Hello, World!"

print(f"Name: {name}")
print(f"Message: {message}")
print(f"Name has {len(name)} characters")
```

Sometimes we need to store whether something is true or false.

```python
# Booleans
is_student = True
is_raining = False

print(f"Student? {is_student}")
print(f"Raining? {is_raining}")
print(f"Both? {is_student and is_raining}")
print(f"Either? {is_student or is_raining}")
```

### 2. Working with Strings

Strings are sequences of characters. Think of text as a string of characters - like beads on a string. Each character has a position, and we can manipulate the whole string or individual parts.

```python
# String basics
text = "Python Programming"

print(f"Original: {text}")
print(f"Uppercase: {text.upper()}")
print(f"Lowercase: {text.lower()}")
print(f"First 6 chars: {text[:6]}") # String slicing - getting parts of a string
print(f"Last word: {text.split()[-1]}")
```

We often need to combine text with variables.

```python
# Different ways to format strings
name = "Alice"
age = 25
score = 95.7

# Method 1: f-strings (recommended - most readable)
message1 = f"Hello {name}, you are {age} years old and scored {score}%"
print(f"F-string: {message1}")

# Method 2: .format() method
message2 = "Hello {}, you are {} years old and scored {}%".format(name, age, score)
print(f"Format method: {message2}")

# Method 3: String concatenation (not recommended for complex cases)
message3 = "Hello " + name + ", you are " + str(age) + " years old"
print(f"Concatenation: {message3}")
```

```python
# String methods
sentence = "The quick brown fox jumps over the lazy dog"

print(f"Length: {len(sentence)} characters")
print(f"Word count: {len(sentence.split())} words")
print(f"Contains 'fox'? {'fox' in sentence}")
print(f"Position of 'fox': {sentence.find('fox')}")
print(f"Replace fox: {sentence.replace('fox', 'cat')}")
```

## Why Do We Need Different Data Structures?

Imagine you're organizing a library. You might use:
- **Lists** for books in order on a shelf
- **Dictionaries** for a card catalog (title → location)
- **Sets** for unique genres (no duplicates)

### 3. Lists

Lists store ordered collections of items and allow duplicates.


```python
# Creating and using lists
fruits = ["apple", "banana", "cherry", "date"]

print(f"Fruits: {fruits}")
print(f"First fruit: {fruits[0]}")
print(f"Last fruit: {fruits[-1]}")
print(f"Number of fruits: {len(fruits)}")

# Modifying lists
fruits.append("elderberry") # Add to end
print(f"After append: {fruits}")

fruits.insert(1, "blueberry") # Insert at position 1
print(f"After insert: {fruits}")
```

```python
# List operations
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

print(f"Original: {numbers}")
print(f"Sum: {sum(numbers)}")
print(f"Maximum: {max(numbers)}")
print(f"Minimum: {min(numbers)}")
print(f"Sorted: {sorted(numbers)}")
print(f"Reversed: {list(reversed(numbers))}")
```

### 4. Dictionaries

Dictionaries store key-value pairs for quick lookups. Think of them like address books - you look up information using a **key** (like a name) to get a **value** (like a phone number).

```python
# Creating dictionaries
student = {
    "name": "Alice",
    "age": 20,
    "major": "Computer Science",
    "gpa": 3.8
}

print(f"Student: {student}")
print(f"Name: {student['name']}")
print(f"Major: {student['major']}")

# Adding and updating
student["year"] = "Junior"
student["gpa"] = 3.9

print(f"Updated: {student}")
```

```python
# Dictionary methods
grades = {"Alice": 95, "Bob": 87, "Carol": 92, "David": 88}

print(f"All students: {list(grades.keys())}")
print(f"All grades: {list(grades.values())}")
print(f"Average grade: {sum(grades.values()) / len(grades):.1f}")

# Finding the top student
top_student = max(grades, key=grades.get)
print(f"Top student: {top_student} with {grades[top_student]}")
```

### 5. Sets

Sets store unique values and support mathematical operations.
Sets are like lists, but with two key differences:
1. **No duplicates** - each item appears only once
2. **No order** - items aren't stored in any particular sequence

```python
# Creating sets
colors1 = {"red", "blue", "green"}
colors2 = {"blue", "yellow", "green"}
# Set operations - like math!
print(f"Set 1: {colors1}")
print(f"Set 2: {colors2}")
print(f"Union: {colors1 | colors2}") # Union: everything from both sets
print(f"Intersection: {colors1 & colors2}") # Intersection: only items in both sets
print(f"Difference: {colors1 - colors2}") # Difference: items in set1 but not set2
```

```python
# Duplicates will be removed!
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
unique_numbers = set(numbers)
# Sets are great for finding unique items
print(f"Original: {numbers}")
print(f"Unique: {unique_numbers}")
print(f"Count of unique: {len(unique_numbers)}")
```

## 6. File Operations

Most real data comes from files - spreadsheets, databases, text files, etc. Let's learn how to read and process file data using Python.

Reading from and writing to files lets us work with external data.

```python
# Writing to a file
data = """apple,5
banana,3
cherry,8
date,2"""

with open('fruits.txt', 'w') as file:
    file.write(data)

print("File created!")

# Method 2: If you have your own file, use the upload button
# Click the folder icon (📁) in the left sidebar, then click upload
print("📁 To upload your own files: Click the folder icon → Upload button")
```

```python
# Reading from a file - Three different approaches

# Method 1: Read entire file at once
# Good for: Small files where you need all content
print("=== Reading entire file ===")
with open('fruits.txt', 'r') as file:
   content = file.read()  # Reads everything into one string
print(content)

# Method 2: Read line by line
# Good for: Large files or when you need line numbers
print("\n=== Reading line by line ===")
with open('fruits.txt', 'r') as file:
   for line_number, line in enumerate(file, 1):  # enumerate adds line numbers
       clean_line = line.strip()  # Remove newline characters
       print(f"Line {line_number}: {clean_line}")

# Method 3: Read and process into data structure
# Good for: Structured data (CSV, key-value pairs)
print("\n=== Processing into dictionary ===")
inventory = {}  # Empty dictionary to store our data

with open('fruits.txt', 'r') as file:
   for line in file:
       # Split each line: "apple,5" becomes ["apple", "5"]
       fruit, count = line.strip().split(',')

       # Store in dictionary with count as integer
       inventory[fruit] = int(count)

# Display results
print(f"Inventory: {inventory}")
print(f"Total items: {sum(inventory.values())}")
```

## 7. Putting It All Together

Let's combine everything to build a complete program.

```python
# Example: Simple library system
library = {
    "books": [
        {"title": "Python Basics", "author": "Smith", "year": 2020, "available": True},
        {"title": "Data Science", "author": "Jones", "year": 2021, "available": True},
        {"title": "AI Ethics", "author": "Brown", "year": 2022, "available": False}
    ]
}

# Find available books
available = [book for book in library["books"] if book["available"]]
print(f"Available books: {len(available)}")

# Get all authors
authors = {book["author"] for book in library["books"]}
print(f"Authors: {authors}")
```
