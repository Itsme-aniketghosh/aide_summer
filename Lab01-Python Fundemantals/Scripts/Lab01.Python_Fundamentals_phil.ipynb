## Part 5: Conditionals and Loops

**Objective**: Process data with logic and iteration

````python
# Demonstrate: Conditional logic
citations = [45, 120, 15, 200, 75]

for count in citations:
    if count > 100:
        impact = "High"
    elif count > 50:
        impact = "Medium"
    else:
        impact = "Low"
    print(f"Citations: {count} - Impact: {impact}")
````

## Part 7: File Operations

**Objective**: Read and process external data files

````python
# Demonstrate: Reading from files
# First, create a sample file
sample_data = """Plato,Ancient,Academy
Aristotle,Ancient,Lyceum
Descartes,Modern,Rationalism
Hume,Modern,Empiricism"""

with open("philosophers.txt", "w") as f:
    f.write(sample_data)

# Read and process the file
philosophers = []
with open("philosophers.txt", "r") as f:
    for line in f:
        name, era, school = line.strip().split(",")
        philosophers.append({"name": name, "era": era, "school": school})

print("Loaded philosophers:")
for p in philosophers:
    print(f"- {p['name']} ({p['era']}): {p['school']}")
````

def analyze_research_paper(title, author, year, citations):
    # Format citation
    formatted = f"{author} ({year}). {title}."
    
    # Analyze title
    word_count = len(title.split())
    is_question = "?" in title
    
    # Determine impact
    if citations > 1000:
        impact = "High"
    elif citations > 100:
        impact = "Medium"
    else:

    # Determine era

    return {

        "word_count": word_count,
        "is_question": is_question,
        "impact_level": impact,
        "era": era
    }
