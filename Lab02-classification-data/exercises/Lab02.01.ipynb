## 3. Pandas Data Manipulation (15 min)

### Selection and Filtering

````python
# Extended philosophical dataset
texts_data = {
    'author': ['Plato', 'Aristotle', 'Kant', 'Mill', 'Rawls', 'Nozick'],
    'work': ['Republic', 'Ethics', 'Critique', 'Liberty', 'Justice', 'Anarchy'],
    'year': [-380, -350, 1781, 1859, 1971, 1974],
    'field': ['Political', 'Ethics', 'Metaphysics', 'Political', 'Political', 'Political'],
    'citations': [15000, 12000, 18000, 7000, 11000, 5000],
    'pages': [700, 400, 800, 150, 500, 300]
}

texts_df = pd.DataFrame(texts_data)

# Essential operations
print("Basic info:")
print(f"Shape: {texts_df.shape}")
print(f"Columns: {texts_df.columns.tolist()}")

# Selection
print("\nSelection examples:")
print("Authors:", texts_df['author'].tolist())
print("\nPolitical philosophy works:")
political = texts_df[texts_df['field'] == 'Political']
print(political[['author', 'work']])

# Multiple conditions
print("\nModern highly-cited works (after 1500, >10k citations):")
modern_cited = texts_df[(texts_df['year'] > 1500) & (texts_df['citations'] > 10000)]
print(modern_cited[['author', 'work', 'citations']])
````

### Grouping and Analysis

````python
# Group by field
print("Analysis by field:")
field_stats = texts_df.groupby('field').agg({
    'citations': ['mean', 'sum'],
    'pages': 'mean',
    'author': 'count'
})
print(field_stats)

# Create new variables
texts_df['citations_per_page'] = texts_df['citations'] / texts_df['pages']
texts_df['era'] = texts_df['year'].apply(
    lambda x: 'Ancient' if x < 0 else 'Modern' if x < 1900 else 'Contemporary'
)

print("\nEfficiency ranking (citations per page):")
efficiency = texts_df.nlargest(3, 'citations_per_page')
print(efficiency[['author', 'work', 'citations_per_page']])

print("\nWorks by era:")
era_counts = texts_df['era'].value_counts()
print(era_counts)
````

## 4. Data Preparation Summary (5 min)

### Key Skills for ML (Lab 2.2 Preview)

````python
# Quick data cleaning example
messy_data = {
    'philosopher': ['kant', 'MILL', None, 'aristotle'],
    'influence': [9.2, 8.7, None, 9.5],
    'field': ['ethics', 'Ethics', 'ethics', 'Ethics']
}

messy_df = pd.DataFrame(messy_data)
print("Messy data:")
print(messy_df)

# Clean up
messy_df['philosopher'] = messy_df['philosopher'].str.title().fillna('Unknown')
messy_df['influence'] = messy_df['influence'].fillna(messy_df['influence'].mean())
messy_df['field'] = messy_df['field'].str.title()

# Create dummy variables
field_dummies = pd.get_dummies(messy_df['field'], prefix='field')
clean_df = pd.concat([messy_df, field_dummies], axis=1)

print("\nCleaned for ML:")
print(clean_df)

print("\nReady for Lab 2.2: Machine Learning with this data!")
````
