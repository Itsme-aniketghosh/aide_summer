## 1. Linear Regression: Predicting Influence Scores (20 min)

### Create Philosophical Dataset

````python
# Create philosophical influence dataset
np.random.seed(42)
philosophers_data = {
    'philosopher': ['Plato', 'Aristotle', 'Kant', 'Mill', 'Rawls', 'Nozick', 'Hume', 'Locke', 'Hobbes', 'Rousseau'],
    'citations': [15000, 12000, 18000, 7000, 11000, 5000, 9000, 8000, 6000, 4000],
    'pages_written': [700, 400, 800, 150, 500, 300, 600, 450, 350, 250],
    'years_active': [50, 35, 40, 25, 30, 20, 35, 40, 45, 30],
    'influence_score': [9.5, 9.2, 9.8, 8.7, 9.1, 8.3, 8.9, 8.5, 8.1, 7.8]
}

phil_df = pd.DataFrame(philosophers_data)
print("Philosophical Influence Dataset:")
print(phil_df)

# Prepare for regression
X = phil_df[['citations', 'pages_written', 'years_active']].values
y = phil_df['influence_score'].values

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"\nTraining set: {X_train.shape[0]} philosophers")
print(f"Test set: {X_test.shape[0]} philosophers")
````

### Simple Linear Regression

````python
# Simple regression: citations -> influence
citations_train = X_train[:, 0].reshape(-1, 1)
citations_test = X_test[:, 0].reshape(-1, 1)

simple_reg = LinearRegression()
simple_reg.fit(citations_train, y_train)

y_pred_simple = simple_reg.predict(citations_test)

print(f"Simple Regression (Citations -> Influence):")
print(f"Coefficient: {simple_reg.coef_[0]:.6f}")
print(f"Intercept: {simple_reg.intercept_:.2f}")
print(f"Test MSE: {mean_squared_error(y_test, y_pred_simple):.4f}")

# Visualize
plt.figure(figsize=(10, 6))
plt.scatter(citations_test, y_test, color='blue', label='Actual', alpha=0.7)
plt.plot(citations_test, y_pred_simple, color='red', label='Predicted', linewidth=2)
plt.xlabel('Citations')
plt.ylabel('Influence Score')
plt.title('Simple Linear Regression: Citations vs Influence')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

### Multiple Linear Regression

````python
# Multiple regression: all features -> influence
# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

multi_reg = LinearRegression()
multi_reg.fit(X_train_scaled, y_train)

y_pred_multi = multi_reg.predict(X_test_scaled)

print(f"Multiple Regression (All Features -> Influence):")
print(f"Coefficients: {multi_reg.coef_}")
print(f"Intercept: {multi_reg.intercept_:.2f}")
print(f"Test MSE: {mean_squared_error(y_test, y_pred_multi):.4f}")

# Feature importance
features = ['Citations', 'Pages Written', 'Years Active']
plt.figure(figsize=(10, 6))
plt.barh(features, multi_reg.coef_)
plt.xlabel('Coefficient Value')
plt.title('Feature Importance in Predicting Philosophical Influence')
plt.grid(True, alpha=0.3)
plt.show()

# Compare models
print(f"\nModel Comparison:")
print(f"Simple Regression MSE: {mean_squared_error(y_test, y_pred_simple):.4f}")
print(f"Multiple Regression MSE: {mean_squared_error(y_test, y_pred_multi):.4f}")
````

## 2. Logistic Regression: Moral Framework Classification (20 min)

### Create Moral Psychology Dataset

````python
# Create moral framework classification dataset
np.random.seed(42)
moral_data = {
    'participant_id': range(1, 101),
    'age': np.random.randint(18, 65, 100),
    'education': np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], 100),
    'trolley_confidence': np.random.randint(1, 11, 100),
    'consequentialist_score': np.random.randint(1, 8, 100),
    'deontological_score': np.random.randint(1, 8, 100),
    'virtue_score': np.random.randint(1, 8, 100)
}

# Create realistic moral framework labels based on scores
frameworks = []
for i in range(100):
    scores = [moral_data['consequentialist_score'][i], 
              moral_data['deontological_score'][i], 
              moral_data['virtue_score'][i]]
    max_idx = np.argmax(scores)
    frameworks.append(['Utilitarian', 'Deontological', 'Virtue Ethics'][max_idx])

moral_data['framework'] = frameworks
moral_df = pd.DataFrame(moral_data)

print("Moral Framework Dataset:")
print(moral_df.head())
print(f"\nFramework distribution:")
print(moral_df['framework'].value_counts())
````

### Logistic Regression Classification

````python
# Prepare data for logistic regression
# Encode education
label_encoder = LabelEncoder()
moral_df['education_encoded'] = label_encoder.fit_transform(moral_df['education'])

# Features and target
X_moral = moral_df[['age', 'education_encoded', 'trolley_confidence', 
                   'consequentialist_score', 'deontological_score', 'virtue_score']].values
y_moral = moral_df['framework'].values

# Split data
X_train_moral, X_test_moral, y_train_moral, y_test_moral = train_test_split(
    X_moral, y_moral, test_size=0.3, random_state=42
)

# Standardize features
scaler_moral = StandardScaler()
X_train_moral_scaled = scaler_moral.fit_transform(X_train_moral)
X_test_moral_scaled = scaler_moral.transform(X_test_moral)

# Train logistic regression
log_reg = LogisticRegression(random_state=42, max_iter=1000)
log_reg.fit(X_train_moral_scaled, y_train_moral)

# Make predictions
y_pred_moral = log_reg.predict(X_test_moral_scaled)
y_pred_proba = log_reg.predict_proba(X_test_moral_scaled)

# Evaluate
accuracy = accuracy_score(y_test_moral, y_pred_moral)
log_loss_val = log_loss(y_test_moral, y_pred_proba)

print(f"Moral Framework Classification Results:")
print(f"Accuracy: {accuracy:.3f}")
print(f"Log Loss: {log_loss_val:.3f}")

# Feature importance
features = ['Age', 'Education', 'Trolley Confidence', 'Consequentialist', 'Deontological', 'Virtue']
plt.figure(figsize=(12, 8))
for i, framework in enumerate(log_reg.classes_):
    plt.subplot(2, 2, i+1)
    plt.barh(features, log_reg.coef_[i])
    plt.title(f'{framework} Coefficients')
    plt.xlabel('Coefficient Value')
    plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

## 3. Decision Trees: Philosophical Argument Analysis (20 min)

### Create Argument Classification Dataset

````python
# Create philosophical argument classification dataset
np.random.seed(42)
argument_data = {
    'argument_id': range(1, 151),
    'premise_count': np.random.randint(1, 6, 150),
    'logical_structure': np.random.choice(['Deductive', 'Inductive', 'Abductive'], 150),
    'complexity_score': np.random.randint(1, 11, 150),
    'evidence_strength': np.random.randint(1, 8, 150),
    'counterarguments': np.random.randint(0, 5, 150)
}

# Create realistic validity labels
validity = []
for i in range(150):
    # Higher evidence strength and fewer counterarguments = more likely valid
    score = (argument_data['evidence_strength'][i] * 0.3 + 
             (5 - argument_data['counterarguments'][i]) * 0.2 +
             (argument_data['premise_count'][i] > 2) * 0.1 +
             np.random.random() * 0.4)
    validity.append('Valid' if score > 0.6 else 'Invalid')

argument_data['validity'] = validity
arg_df = pd.DataFrame(argument_data)

print("Philosophical Argument Dataset:")
print(arg_df.head())
print(f"\nValidity distribution:")
print(arg_df['validity'].value_counts())
````

### Decision Tree Classification

````python
# Prepare data for decision tree
# Encode logical structure
structure_encoder = LabelEncoder()
arg_df['structure_encoded'] = structure_encoder.fit_transform(arg_df['logical_structure'])

# Features and target
X_arg = arg_df[['premise_count', 'structure_encoded', 'complexity_score', 
               'evidence_strength', 'counterarguments']].values
y_arg = arg_df['validity'].values

# Split data
X_train_arg, X_test_arg, y_train_arg, y_test_arg = train_test_split(
    X_arg, y_arg, test_size=0.3, random_state=42
)

# Train decision tree
tree_clf = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=42)
tree_clf.fit(X_train_arg, y_train_arg)

# Make predictions
y_pred_tree = tree_clf.predict(X_test_arg)

# Evaluate
tree_accuracy = accuracy_score(y_test_arg, y_pred_tree)
print(f"Decision Tree Classification Results:")
print(f"Accuracy: {tree_accuracy:.3f}")

# Feature importance
feature_names = ['Premise Count', 'Logical Structure', 'Complexity', 'Evidence Strength', 'Counterarguments']
importances = tree_clf.feature_importances_

plt.figure(figsize=(10, 6))
plt.barh(feature_names, importances)
plt.xlabel('Feature Importance')
plt.title('Feature Importance in Argument Validity Prediction')
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nFeature Importance Ranking:")
for i, (feature, importance) in enumerate(zip(feature_names, importances)):
    print(f"{i+1}. {feature}: {importance:.3f}")
````

### Visualize Decision Tree

````python
# Visualize the decision tree
plt.figure(figsize=(15, 10))
plot_tree(tree_clf, 
          feature_names=feature_names,
          class_names=['Invalid', 'Valid'],
          filled=True, 
          fontsize=10)
plt.title('Decision Tree for Philosophical Argument Validity')
plt.show()

# Interpret decision rules
print("\nDecision Rules Interpretation:")
print("The tree shows how to classify argument validity based on:")
print("1. Evidence strength (most important)")
print("2. Number of counterarguments")
print("3. Premise count and complexity")
print("4. Logical structure type")
````
