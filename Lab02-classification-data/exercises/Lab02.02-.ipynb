## 1. Linear Regression - Predicting Numbers (25 min)

### What is Linear Regression?
Linear regression finds the best line through your data points. It's like drawing a trend line to predict future values.

**Intuition**: If you know someone's study hours, can you predict their test score? Linear regression finds the relationship.

### Simple Example

````python
# Create simple research dataset
data = {
    'researcher': ['Alice', 'Bob', 'Carol', 'David', 'Eve', 'Frank'],
    'publications': [5, 10, 15, 20, 25, 30],
    'citations': [50, 120, 200, 300, 400, 500],
    'influence_score': [6.2, 7.1, 7.8, 8.5, 8.9, 9.2]
}

df = pd.DataFrame(data)
print("Research Dataset:")
print(df)

# Visualize the relationship
plt.figure(figsize=(10, 4))

# Plot 1: Publications vs Influence
plt.subplot(1, 2, 1)
plt.scatter(df['publications'], df['influence_score'], color='blue')
plt.xlabel('Publications')
plt.ylabel('Influence Score')
plt.title('Publications vs Influence')
plt.grid(True, alpha=0.3)

# Plot 2: Citations vs Influence
plt.subplot(1, 2, 2)
plt.scatter(df['citations'], df['influence_score'], color='red')
plt.xlabel('Citations')
plt.ylabel('Influence Score')
plt.title('Citations vs Influence')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nObservation: Both show upward trends - good for linear regression!")
````

### Step-by-Step: Simple Linear Regression

**Goal**: Predict influence score from number of publications

````python
# Step 1: Prepare the data
X = df[['publications']].values  # Features (input) - note the [[ ]] for 2D array
y = df['influence_score'].values  # Target (output)

print("Step 1: Data prepared")
print(f"X shape: {X.shape} (samples, features)")
print(f"y shape: {y.shape} (samples,)")

# Step 2: Split data into training and testing sets
# Why? We need to test our model on unseen data to see if it generalizes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"\nStep 2: Data split")
print(f"Training set: {len(X_train)} samples")
print(f"Test set: {len(X_test)} samples")

# Step 3: Create and train the model
model = LinearRegression()  # Create the model
model.fit(X_train, y_train)  # Train it on training data

print(f"\nStep 3: Model trained")
print(f"Slope (coefficient): {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
print(f"Equation: influence = {model.coef_[0]:.4f} * publications + {model.intercept_:.4f}")

# Step 4: Make predictions
y_pred = model.predict(X_test)

print(f"\nStep 4: Predictions made")
print(f"Actual values: {y_test}")
print(f"Predicted values: {y_pred}")

# Step 5: Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"\nStep 5: Model evaluation")
print(f"Mean Squared Error: {mse:.4f}")
print(f"Lower MSE = better model")
````

### Visualize the Results

````python
# Plot the regression line
plt.figure(figsize=(10, 6))

# Plot all data points
plt.scatter(X_train, y_train, color='blue', label='Training Data', alpha=0.7)
plt.scatter(X_test, y_test, color='red', label='Test Data', alpha=0.7)

# Plot the regression line
# Create a range of x values for smooth line
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_line = model.predict(x_range)
plt.plot(x_range, y_line, color='green', linewidth=2, label='Regression Line')

# Add predictions
plt.scatter(X_test, y_pred, color='orange', marker='x', s=100, label='Predictions')

plt.xlabel('Publications')
plt.ylabel('Influence Score')
plt.title('Linear Regression: Publications → Influence Score')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("Interpretation:")
print(f"- For every additional publication, influence score increases by {model.coef_[0]:.4f}")
print(f"- Green line shows the model's predictions")
print(f"- Orange X marks show predictions for test data")
````

### Multiple Linear Regression

**Why Multiple Features?** 
Real-world predictions usually depend on multiple factors. Instead of just publications, let's use both publications AND citations.

**Intuition**: Like predicting house price from both size AND location, not just size.

````python
# Multiple regression: Use both publications and citations
print("=== MULTIPLE LINEAR REGRESSION ===")

# Step 1: Prepare data with multiple features
X_multi = df[['publications', 'citations']].values  # Two features now
y_multi = df['influence_score'].values

print(f"Multiple features shape: {X_multi.shape}")
print(f"Features: publications, citations")

# Step 2: Split data
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(
    X_multi, y_multi, test_size=0.3, random_state=42
)

# Step 3: Scale features (important for multiple regression)
# Why? Features have different scales (publications: 5-30, citations: 50-500)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_multi)  # Fit on training data
X_test_scaled = scaler.transform(X_test_multi)        # Transform test data

print(f"\nBefore scaling - Training data:")
print(f"Publications range: {X_train_multi[:, 0].min():.1f} to {X_train_multi[:, 0].max():.1f}")
print(f"Citations range: {X_train_multi[:, 1].min():.1f} to {X_train_multi[:, 1].max():.1f}")

print(f"\nAfter scaling - Training data:")
print(f"Publications range: {X_train_scaled[:, 0].min():.1f} to {X_train_scaled[:, 0].max():.1f}")
print(f"Citations range: {X_train_scaled[:, 1].min():.1f} to {X_train_scaled[:, 1].max():.1f}")

# Step 4: Train model
multi_model = LinearRegression()
multi_model.fit(X_train_scaled, y_train_multi)

# Step 5: Make predictions
y_pred_multi = multi_model.predict(X_test_scaled)

# Step 6: Evaluate
mse_multi = mean_squared_error(y_test_multi, y_pred_multi)

print(f"\nMultiple Regression Results:")
print(f"MSE: {mse_multi:.4f}")
print(f"Coefficients: {multi_model.coef_}")
print(f"Intercept: {multi_model.intercept_:.4f}")

# Feature importance visualization
features = ['Publications', 'Citations']
plt.figure(figsize=(8, 5))
plt.bar(features, multi_model.coef_, color=['blue', 'red'])
plt.xlabel('Features')
plt.ylabel('Coefficient Value')
plt.title('Feature Importance in Multiple Regression')
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nInterpretation:")
print(f"- Coefficient for publications: {multi_model.coef_[0]:.4f}")
print(f"- Coefficient for citations: {multi_model.coef_[1]:.4f}")
print(f"- Larger coefficient = more important feature")
````

## 2. Logistic Regression - Predicting Categories (25 min)

### What is Logistic Regression?
While linear regression predicts numbers, logistic regression predicts categories (classes).

**Intuition**: Instead of "How much will it rain?", we ask "Will it rain or not?" (Yes/No)

**Key Difference**: Output is probability (0 to 1), not a continuous number.

### Simple Classification Example

````python
# Create a simple classification dataset
# Goal: Predict if someone will choose Option A or B based on their age and education

np.random.seed(42)
classification_data = {
    'age': [22, 35, 45, 28, 55, 31, 40, 25, 50, 33, 38, 29],
    'education_years': [16, 18, 20, 17, 22, 19, 21, 16, 23, 18, 20, 17],
    'choice': ['A', 'B', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A']
}

class_df = pd.DataFrame(classification_data)
print("Classification Dataset:")
print(class_df)

# Visualize the data
plt.figure(figsize=(8, 6))

# Plot points with different colors for each choice
choice_a = class_df[class_df['choice'] == 'A']
choice_b = class_df[class_df['choice'] == 'B']

plt.scatter(choice_a['age'], choice_a['education_years'], 
           color='red', label='Choice A', s=100)
plt.scatter(choice_b['age'], choice_b['education_years'], 
           color='blue', label='Choice B', s=100)

plt.xlabel('Age')
plt.ylabel('Education Years')
plt.title('Choice Distribution by Age and Education')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print("\nPattern: Can we predict choice from age and education?")
print(f"Choice A count: {len(choice_a)}")
print(f"Choice B count: {len(choice_b)}")
````

### Step-by-Step: Logistic Regression

````python
print("=== LOGISTIC REGRESSION STEP-BY-STEP ===")

# Step 1: Prepare data
X_class = class_df[['age', 'education_years']].values
y_class = class_df['choice'].values

print(f"Step 1: Data prepared")
print(f"Features: {X_class.shape}")
print(f"Target classes: {np.unique(y_class)}")

# Step 2: Split data
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X_class, y_class, test_size=0.3, random_state=42
)

print(f"\nStep 2: Data split")
print(f"Training samples: {len(X_train_class)}")
print(f"Test samples: {len(X_test_class)}")

# Step 3: Scale features (important for logistic regression)
scaler_class = StandardScaler()
X_train_class_scaled = scaler_class.fit_transform(X_train_class)
X_test_class_scaled = scaler_class.transform(X_test_class)

print(f"\nStep 3: Features scaled")

# Step 4: Train logistic regression
log_model = LogisticRegression(random_state=42)
log_model.fit(X_train_class_scaled, y_train_class)

print(f"\nStep 4: Model trained")
print(f"Classes learned: {log_model.classes_}")

# Step 5: Make predictions
y_pred_class = log_model.predict(X_test_class_scaled)
y_pred_proba = log_model.predict_proba(X_test_class_scaled)

print(f"\nStep 5: Predictions made")
print(f"Actual: {y_test_class}")
print(f"Predicted: {y_pred_class}")
print(f"\nPrediction probabilities:")
for i, (actual, pred, prob) in enumerate(zip(y_test_class, y_pred_class, y_pred_proba)):
    print(f"Sample {i+1}: Actual={actual}, Predicted={pred}, Prob(A)={prob[0]:.3f}, Prob(B)={prob[1]:.3f}")

# Step 6: Evaluate
accuracy = accuracy_score(y_test_class, y_pred_class)
print(f"\nStep 6: Model evaluation")
print(f"Accuracy: {accuracy:.3f} ({accuracy*100:.1f}%)")
print(f"This means {accuracy*100:.1f}% of predictions were correct")
````

## 3. Decision Trees - Interpretable Decisions (20 min)

### What is a Decision Tree?
A decision tree makes decisions by asking a series of yes/no questions, like a flowchart.

**Intuition**: Like asking "Is it sunny?" → If yes: "Go to beach", If no: "Stay home"

**Advantage**: Easy to understand and explain (unlike black-box models)

### Simple Decision Tree Example

````python
# Create a simple decision tree dataset
# Goal: Predict if someone will buy a product based on age, income, and interest

np.random.seed(42)
tree_data = {
    'age': [25, 35, 45, 28, 55, 31, 40, 25, 50, 33, 38, 29, 22, 60, 42, 30],
    'income': [30, 50, 70, 35, 80, 45, 65, 32, 75, 48, 62, 38, 28, 85, 68, 42],  # in thousands
    'interest_score': [7, 8, 6, 9, 5, 8, 7, 9, 4, 8, 6, 9, 8, 3, 5, 7],  # 1-10 scale
    'will_buy': ['Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes']
}

tree_df = pd.DataFrame(tree_data)
print("Decision Tree Dataset:")
print(tree_df)

print(f"\nPurchase distribution:")
print(tree_df['will_buy'].value_counts())

# Quick visualization
plt.figure(figsize=(12, 4))

# Plot 1: Age vs Income
plt.subplot(1, 3, 1)
buyers = tree_df[tree_df['will_buy'] == 'Yes']
non_buyers = tree_df[tree_df['will_buy'] == 'No']

plt.scatter(buyers['age'], buyers['income'], color='green', label='Will Buy', s=60)
plt.scatter(non_buyers['age'], non_buyers['income'], color='red', label='Won\'t Buy', s=60)
plt.xlabel('Age')
plt.ylabel('Income (k)')
plt.title('Age vs Income')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: Interest Score distribution
plt.subplot(1, 3, 2)
plt.hist(buyers['interest_score'], alpha=0.7, label='Will Buy', color='green', bins=5)
plt.hist(non_buyers['interest_score'], alpha=0.7, label='Won\'t Buy', color='red', bins=5)
plt.xlabel('Interest Score')
plt.ylabel('Count')
plt.title('Interest Score Distribution')
plt.legend()

# Plot 3: Summary
plt.subplot(1, 3, 3)
purchase_counts = tree_df['will_buy'].value_counts()
plt.bar(purchase_counts.index, purchase_counts.values, color=['red', 'green'])
plt.xlabel('Decision')
plt.ylabel('Count')
plt.title('Purchase Decision Distribution')

plt.tight_layout()
plt.show()

print("\nObservation: Look for patterns that might help predict purchases")
````

### Step-by-Step: Decision Tree

````python
print("=== DECISION TREE STEP-BY-STEP ===")

# Step 1: Prepare data
X_tree = tree_df[['age', 'income', 'interest_score']].values
y_tree = tree_df['will_buy'].values

print(f"Step 1: Data prepared")
print(f"Features: {X_tree.shape}")
print(f"Classes: {np.unique(y_tree)}")

# Step 2: Split data
X_train_tree, X_test_tree, y_train_tree, y_test_tree = train_test_split(
    X_tree, y_tree, test_size=0.3, random_state=42
)

print(f"\nStep 2: Data split")
print(f"Training samples: {len(X_train_tree)}")
print(f"Test samples: {len(X_test_tree)}")

# Step 3: Train decision tree
# Note: No scaling needed for decision trees!
tree_model = DecisionTreeClassifier(
    max_depth=3,           # Limit depth to avoid overfitting
    random_state=42,       # For reproducible results
    criterion='entropy'    # How to measure "impurity"
)

tree_model.fit(X_train_tree, y_train_tree)

print(f"\nStep 3: Decision tree trained")
print(f"Tree depth: {tree_model.get_depth()}")
print(f"Number of leaves: {tree_model.get_n_leaves()}")

# Step 4: Make predictions
y_pred_tree = tree_model.predict(X_test_tree)
y_pred_proba_tree = tree_model.predict_proba(X_test_tree)

print(f"\nStep 4: Predictions made")
print(f"Actual: {y_test_tree}")
print(f"Predicted: {y_pred_tree}")

# Step 5: Evaluate
accuracy_tree = accuracy_score(y_test_tree, y_pred_tree)
print(f"\nStep 5: Model evaluation")
print(f"Accuracy: {accuracy_tree:.3f} ({accuracy_tree*100:.1f}%)")

# Step 6: Feature importance
feature_names = ['Age', 'Income', 'Interest Score']
importances = tree_model.feature_importances_

print(f"\nStep 6: Feature importance")
for feature, importance in zip(feature_names, importances):
    print(f"{feature}: {importance:.3f}")

# Visualize feature importance
plt.figure(figsize=(8, 5))
plt.bar(feature_names, importances, color=['blue', 'green', 'red'])
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importance in Decision Tree')
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nInterpretation:")
print(f"- Higher importance = more useful for making decisions")
print(f"- Most important feature: {feature_names[np.argmax(importances)]}")
````

### Visualize the Decision Tree

**This is the magic of decision trees - you can see exactly how they make decisions!**

````python
# Visualize the decision tree
plt.figure(figsize=(15, 10))

plot_tree(tree_model, 
          feature_names=feature_names,
          class_names=['No', 'Yes'],  # Alphabetical order
          filled=True,                # Color the nodes
          fontsize=10,
          rounded=True)

plt.title('Decision Tree for Purchase Prediction')
plt.show()

print("\nHow to read the tree:")
print("1. Start at the top (root)")
print("2. Follow the path based on yes/no answers")
print("3. Reach a leaf node for the final prediction")
print("\nExample path:")
print("- If Interest Score > 5.5 AND Age > 29.5 → Predict 'Yes'")
print("- If Interest Score ≤ 5.5 → Predict 'No'")

# Show some prediction examples
print("\n=== PREDICTION EXAMPLES ===")
test_cases = [
    [30, 50, 8],  # Age=30, Income=50k, Interest=8
    [25, 40, 4],  # Age=25, Income=40k, Interest=4
    [45, 70, 7],  # Age=45, Income=70k, Interest=7
]

for i, case in enumerate(test_cases):
    prediction = tree_model.predict([case])[0]
    probability = tree_model.predict_proba([case])[0]
    print(f"Case {i+1}: Age={case[0]}, Income={case[1]}k, Interest={case[2]}")
    print(f"  Prediction: {prediction}")
    print(f"  Probability: {probability[1]:.3f} for 'Yes'")
    print()
````

### What You've Learned

#### 1. **Linear Regression** (Predicting Numbers)
- **Use when**: You want to predict a continuous number
- **How it works**: Finds the best line through your data points
- **Evaluation**: Mean Squared Error (MSE) - lower is better
- **Example**: Predict house price from size and location

#### 2. **Logistic Regression** (Predicting Categories)
- **Use when**: You want to predict categories (Yes/No, A/B/C)
- **How it works**: Calculates probability of each category
- **Evaluation**: Accuracy - higher is better
- **Example**: Predict if email is spam or not

#### 3. **Decision Trees** (Interpretable Decisions)
- **Use when**: You need to explain how decisions are made
- **How it works**: Asks yes/no questions to reach a decision
- **Evaluation**: Accuracy + feature importance
- **Example**: Medical diagnosis with clear reasoning

### Essential Code Pattern
```python
# 1. Prepare data
X = data[['feature1', 'feature2']].values
y = data['target'].values

# 2. Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# 3. Scale features (for regression/logistic)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 4. Train model
model = LinearRegression()  # or LogisticRegression() or DecisionTreeClassifier()
model.fit(X_train_scaled, y_train)

# 5. Make predictions
predictions = model.predict(X_test_scaled)

# 6. Evaluate
score = mean_squared_error(y_test, predictions)  # or accuracy_score()
```

### When to Use Each Algorithm

| Algorithm | Problem Type | Output | Interpretability | Use Case |
|-----------|-------------|--------|------------------|----------|
| **Linear Regression** | Predict numbers | Continuous value | Medium | Sales forecasting |
| **Logistic Regression** | Predict categories | Probability + class | Medium | Medical diagnosis |
| **Decision Trees** | Predict categories | Class + reasoning | High | Rule-based decisions |
