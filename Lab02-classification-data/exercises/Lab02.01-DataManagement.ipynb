### Row Selection and Indexing

Two main methods for selecting rows:
- **iloc**: Integer-based indexing (position)
- **loc**: Label-based indexing (names/labels)

````python
print("=== ROW SELECTION METHODS ===")

# 1. iloc - integer position based
print("\n1. ILOC - INTEGER POSITION:")
print("First row (index 0):")
print(df.iloc[0])

print("\nFirst 3 rows:")
print(df.iloc[0:3])

print("\nSpecific cell - Row 1, Column 2:")
print(df.iloc[1, 2])  # Should be Kant's year

# 2. loc - label based (using index labels)
print("\n2. LOC - LABEL BASED:")
print("Row with index 2:")
print(df.loc[2])

print("\nSpecific cell by label:")
print(df.loc[2, 'Author'])  # Should be Kant

# 3. Create custom index for more meaningful access
print("\n3. CUSTOM INDEX:")
df_custom = df.set_index('Author')
print("\nKant's information:")
print(df_custom.loc['Kant'])

print("\nKant's work:")
print(df_custom.loc['Kant', 'Work'])
````

### Slicing and Advanced Selection

````python
print("=== SLICING TECHNIQUES ===")

# 1. Row slicing
print("\n1. ROW SLICING:")
print("Rows 1-3:")
print(df.iloc[1:4])

# 2. Column slicing
print("\n2. COLUMN SLICING:")
print("Columns from Author to Field:")
print(df.loc[:, 'Author':'Field'])

# 3. Combined row and column slicing
print("\n3. COMBINED SLICING:")
print("First 3 rows, specific columns:")
print(df.iloc[0:3, 1:4])

# 4. Using custom index
print("\n4. SLICING WITH CUSTOM INDEX:")
df_custom = df.set_index('Author')
print("From Plato to Kant:")
print(df_custom.loc['Plato':'Kant', 'Work':'Field'])
````

### Filtering Data Based on Conditions

This is crucial for philosophical research - finding texts that meet specific criteria.

````python
print("=== FILTERING PHILOSOPHICAL DATA ===")

# 1. Simple conditions
print("\n1. WORKS BY FIELD:")
ethics_works = df[df['Field'] == 'Ethics']
print(f"Ethics works: {len(ethics_works)}")
print(ethics_works[['Author', 'Work', 'Year']])

# 2. Numerical conditions
print("\n2. HIGHLY CITED WORKS (>8000 citations):")
highly_cited = df[df['Citations'] > 8000]
print(highly_cited[['Author', 'Work', 'Citations']])

# 3. Multiple conditions with & (and)
print("\n3. MODERN ETHICS (after 1800, Ethics field):")
modern_ethics = df[(df['Year'] > 1800) & (df['Field'] == 'Ethics')]
print(modern_ethics[['Author', 'Work', 'Year']])

# 4. Multiple conditions with | (or)
print("\n4. ANCIENT OR HIGHLY INFLUENTIAL (before 0 OR influence > 9):")
ancient_or_influential = df[(df['Year'] < 0) | (df['Influence_Score'] > 9)]
print(ancient_or_influential[['Author', 'Work', 'Year', 'Influence_Score']])

# 5. String operations
print("\n5. WORKS WITH 'ETHICS' IN TITLE:")
ethics_titles = df[df['Work'].str.contains('Ethics', case=False, na=False)]
print(ethics_titles[['Author', 'Work']])

# 6. Filtering by multiple values
print("\n6. WORKS BY SPECIFIC AUTHORS:")
selected_authors = df[df['Author'].isin(['Plato', 'Aristotle', 'Kant'])]
print(selected_authors[['Author', 'Work', 'Year']])
````

## 4. Data Manipulation and Analysis (25 min)

### Sorting and Ranking

````python
print("=== SORTING AND RANKING ===")

# 1. Sort by single column
print("\n1. WORKS BY CHRONOLOGICAL ORDER:")
chronological = df.sort_values('Year')
print(chronological[['Author', 'Work', 'Year']])

# 2. Sort by multiple columns
print("\n2. SORT BY FIELD, THEN BY CITATIONS:")
field_citations = df.sort_values(['Field', 'Citations'], ascending=[True, False])
print(field_citations[['Field', 'Author', 'Citations']])

# 3. Top N analysis
print("\n3. TOP 3 MOST CITED WORKS:")
top_cited = df.nlargest(3, 'Citations')
print(top_cited[['Author', 'Work', 'Citations']])

print("\n4. LEAST CITED WORKS:")
least_cited = df.nsmallest(2, 'Citations')
print(least_cited[['Author', 'Work', 'Citations']])
````

### Grouping and Aggregation

Essential for analyzing patterns across categories in philosophical data.

````python
print("=== GROUPING AND AGGREGATION ===")

# 1. Group by field
print("\n1. ANALYSIS BY PHILOSOPHICAL FIELD:")
field_analysis = df.groupby('Field').agg({
    'Citations': ['mean', 'sum', 'count'],
    'Influence_Score': 'mean',
    'Pages': 'mean'
})
print(field_analysis.round(2))

# 2. Create time periods and group by them
print("\n2. ANALYSIS BY TIME PERIOD:")
def categorize_period(year):
    if year < 0:
        return 'Ancient'
    elif year < 1500:
        return 'Medieval'
    elif year < 1900:
        return 'Modern'
    else:
        return 'Contemporary'

df['Period'] = df['Year'].apply(categorize_period)
period_analysis = df.groupby('Period').agg({
    'Citations': 'mean',
    'Influence_Score': 'mean',
    'Author': 'count'
})
print(period_analysis.round(2))

# 3. Cross-tabulation
print("\n3. WORKS BY FIELD AND PERIOD:")
cross_tab = pd.crosstab(df['Field'], df['Period'])
print(cross_tab)
````

### Creating New Variables

Often you'll need to create new variables for analysis.

````python
print("=== CREATING NEW VARIABLES ===")

# 1. Simple calculations
print("\n1. CITATIONS PER PAGE:")
df['Citations_Per_Page'] = df['Citations'] / df['Pages']
print(df[['Author', 'Work', 'Citations_Per_Page']].round(2))

# 2. Conditional variables
print("\n2. INFLUENCE CATEGORIES:")
df['Influence_Category'] = df['Influence_Score'].apply(
    lambda x: 'Highly Influential' if x >= 9.0 else 'Moderately Influential' if x >= 8.0 else 'Less Influential'
)
print(df[['Author', 'Work', 'Influence_Score', 'Influence_Category']])

# 3. Text processing
print("\n3. WORK LENGTH CATEGORIES:")
df['Length_Category'] = pd.cut(df['Pages'], 
                              bins=[0, 100, 400, 1000], 
                              labels=['Short', 'Medium', 'Long'])
print(df[['Author', 'Work', 'Pages', 'Length_Category']])

# 4. Complex conditions
print("\n4. ACCESSIBILITY SCORE:")
# Combine multiple factors: shorter works with high influence are more accessible
df['Accessibility_Score'] = (
    df['Influence_Score'] * 0.7 +  # Influence weight
    (1000 - df['Pages']) / 1000 * 0.3  # Length weight (shorter = more accessible)
) * 10
print(df[['Author', 'Work', 'Accessibility_Score']].round(2))
````

## 5. Preparing Data for Machine Learning (15 min)

### Data Cleaning and Preparation

Before applying ML algorithms, data must be cleaned and formatted properly.

````python
print("=== DATA PREPARATION FOR ML ===")

# Create a dataset with some common issues
messy_data = {
    'philosopher': ['plato', 'ARISTOTLE', 'Kant', 'mill', 'rawls', None, 'Nozick'],
    'birth_year': [-428, -384, 1724, 1806, 1921, 1938, 1938],
    'field': ['political', 'Ethics', 'ethics', 'Ethics', 'political', 'Political', 'political'],
    'influence': [9.8, 9.5, 9.2, 8.7, 9.1, None, 8.3],
    'major_work': ['Republic', 'Nicomachean Ethics', 'Groundwork', 'Utilitarianism', 
                  'Theory of Justice', 'Anarchy State Utopia', 'Anarchy, State, and Utopia']
}

messy_df = pd.DataFrame(messy_data)
print("Original messy data:")
print(messy_df)
print("\nData info:")
print(messy_df.info())

# 1. Handle missing values
print("\n1. HANDLING MISSING VALUES:")
print("Missing values per column:")
print(messy_df.isnull().sum())

# Fill missing philosopher name
messy_df['philosopher'] = messy_df['philosopher'].fillna('Unknown')

# Fill missing influence with median
messy_df['influence'] = messy_df['influence'].fillna(messy_df['influence'].median())

# 2. Standardize text data
print("\n2. STANDARDIZING TEXT DATA:")
# Standardize philosopher names
messy_df['philosopher'] = messy_df['philosopher'].str.title()

# Standardize field names
field_mapping = {
    'political': 'Political Philosophy',
    'Political': 'Political Philosophy',
    'ethics': 'Ethics',
    'Ethics': 'Ethics'
}
messy_df['field'] = messy_df['field'].map(field_mapping)

print("Cleaned data:")
print(messy_df)

# 3. Create categorical variables for ML
print("\n3. ENCODING CATEGORICAL VARIABLES:")
# One-hot encoding for field
field_dummies = pd.get_dummies(messy_df['field'], prefix='field')
print("Field dummy variables:")
print(field_dummies)

# Combine with original data
ml_ready_df = pd.concat([messy_df, field_dummies], axis=1)
print("\nML-ready dataset:")
print(ml_ready_df[['philosopher', 'birth_year', 'influence', 'field_Ethics', 'field_Political Philosophy']])
````

### Feature Engineering for Philosophical Data

````python
print("=== FEATURE ENGINEERING ===")

# Working with our original philosophical texts data
ml_df = texts_df.copy()

# 1. Create numerical features
print("\n1. NUMERICAL FEATURES:")
ml_df['citations_per_page'] = ml_df['Citations'] / ml_df['Pages']
ml_df['age_of_work'] = 2024 - ml_df['Year']
ml_df['is_ancient'] = (ml_df['Year'] < 0).astype(int)
ml_df['is_modern'] = ((ml_df['Year'] >= 1500) & (ml_df['Year'] < 1900)).astype(int)

# 2. Categorical encoding
print("\n2. CATEGORICAL ENCODING:")
field_encoded = pd.get_dummies(ml_df['Field'], prefix='field')
ml_df = pd.concat([ml_df, field_encoded], axis=1)

# 3. Scaling numerical features
print("\n3. FEATURE SCALING:")
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Select numerical features to scale
numerical_features = ['Citations', 'Pages', 'Influence_Score', 'citations_per_page', 'age_of_work']
ml_df[numerical_features] = scaler.fit_transform(ml_df[numerical_features])

# 4. Final ML dataset
print("\n4. FINAL ML DATASET:")
ml_features = ['Citations', 'Pages', 'Influence_Score', 'citations_per_page', 'age_of_work', 
               'is_ancient', 'is_modern', 'field_Ethics', 'field_Political Philosophy']
X = ml_df[ml_features]
y = ml_df['Influence_Score']  # Target variable

print(f"Feature matrix shape: {X.shape}")
print(f"Target variable shape: {y.shape}")
print("\nFeatures:")
print(X.head())

print("\nData is ready for machine learning algorithms!")
````
