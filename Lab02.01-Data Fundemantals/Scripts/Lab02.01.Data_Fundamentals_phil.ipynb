````python
# Task 3: Categorize papers by depth
# Create categories based on summary length: Brief (<300), Standard (300-600), Detailed (>600)
# Expected output: Distribution of consciousness papers by depth
# Write your code here:
````

<details>
<summary>💡 Hint for Task 3</summary>

```python
# Get summary lengths from original data
consciousness_subset['summary_length'] = df.loc[consciousness_subset.index, 'summary_length']

# Create research depth categories
consciousness_subset['summary_depth'] = pd.cut(
    consciousness_subset['summary_length'],
    bins=[0, 300, 600, 10000],
    labels=['Brief', 'Standard', 'Detailed']
)

# Show distribution
depth_counts = consciousness_subset['summary_depth'].value_counts()
print("Research depth distribution:")
for depth, count in depth_counts.items():
    print(f"  {depth}: {count} papers ({count/len(consciousness_subset)*100:.1f}%)")
```

plt.figure(figsize=(8, 6))
depth_counts = consciousness_subset['research_depth'].value_counts()

# Create bar chart
bars = plt.bar(depth_counts.index, depth_counts.values, 
                color=['lightcoral', 'skyblue', 'lightgreen'])

# Add value labels on bars

    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{count}\n({count/len(consciousness_subset)*100:.0f}%)',
             ha='center', va='bottom')

plt.title('Consciousness Papers by Research Depth', fontsize=14)
plt.xlabel('Research Depth')
plt.ylabel('Number of Papers')
plt.grid(axis='y', alpha=0.3)
plt.show()
