## 1. Data Structures Review (30 min)

Python offers three essential data structures for organizing information:
- **Lists**: Ordered collections (like a reading list)
- **Dictionaries**: Key-value pairs (like a glossary)
- **Sets**: Unique items only (like philosophical schools)

### Lists: Ordered Collections

```python
# Lists maintain order and allow duplicates
philosophers = ["Plato", "Aristotle", "Descartes", "Kant", "Nietzsche"]

# Basic list operations
print(f"First philosopher: {philosophers[0]}")
print(f"Last philosopher: {philosophers[-1]}")
print(f"Total philosophers: {len(philosophers)}")

# Adding and removing items
philosophers.append("Sartre")  # Add to end
philosophers.insert(2, "Aquinas")  # Insert at position
print(f"\nUpdated list: {philosophers}")

# List slicing
ancient = philosophers[0:2]  # First two
modern = philosophers[3:]    # From index 3 onward
print(f"\nAncient philosophers: {ancient}")
print(f"Modern philosophers: {modern}")
```

### Dictionaries: Key-Value Mappings

```python
# Dictionaries store information with labels
philosopher_info = {
    "name": "Socrates",
    "birth": -470,
    "death": -399,
    "method": "Socratic dialogue",
    "famous_quote": "The unexamined life is not worth living"
}

# Accessing values
print(f"Name: {philosopher_info['name']}")
print(f"Method: {philosopher_info['method']}")

# Adding new information
philosopher_info["students"] = ["Plato", "Xenophon"]
philosopher_info["city"] = "Athens"

# Iterating through dictionary
print("\nPhilosopher details:")
for key, value in philosopher_info.items():
    print(f"  {key}: {value}")
```

### Sets: Unique Collections

```python
# Sets automatically remove duplicates
ethical_concepts = {"virtue", "duty", "utility", "justice", "virtue"}  # Note duplicate
print(f"Unique ethical concepts: {ethical_concepts}")

# Set operations
kantian_ethics = {"duty", "categorical imperative", "autonomy"}
utilitarian_ethics = {"utility", "happiness", "consequences"}

# Finding commonalities and differences
print(f"\nShared concepts: {kantian_ethics & utilitarian_ethics}")
print(f"All concepts: {kantian_ethics | utilitarian_ethics}")
print(f"Unique to Kant: {kantian_ethics - utilitarian_ethics}")
```

## 2. String Processing (25 min)

## 3. Control Flow (25 min)

Control flow structures let us make decisions and repeat operations - essential for data processing.

### Conditionals: Making Decisions

```python
# Simple if-else
year = 1650

if year < 0:
    era = "BCE"
else:
    era = "CE"
print(f"Year {year} is in {era}")

# Multiple conditions with elif
def classify_philosopher_era(birth_year):
    """Classify philosopher by historical period"""
    if birth_year < -600:
        return "Pre-Socratic"
    elif birth_year < 0:
        return "Ancient"
    elif birth_year < 500:
        return "Late Ancient"
    elif birth_year < 1400:
        return "Medieval"
    elif birth_year < 1800:
        return "Modern"
    else:
        return "Contemporary"

# Test the function
philosophers = [
    {"name": "Thales", "birth": -624},
    {"name": "Plato", "birth": -428},
    {"name": "Descartes", "birth": 1596},
    {"name": "Sartre", "birth": 1905}
]

for phil in philosophers:
    era = classify_philosopher_era(phil["birth"])
    print(f"{phil['name']}: {era}")
```

### Loops: Repeating Operations

```python
# For loops - iterate through collections
schools_of_thought = ["Rationalism", "Empiricism", "Idealism", "Pragmatism"]

print("Major philosophical schools:")
for school in schools_of_thought:
    print(f"- {school}")

# Using enumerate for numbering
print("\nNumbered list:")
for i, school in enumerate(schools_of_thought, 1):
    print(f"{i}. {school}")

# While loops - repeat until condition is false
print("\nCounting arguments:")
premises = ["All men are mortal", "Socrates is a man"]
i = 0
while i < len(premises):
    print(f"Premise {i+1}: {premises[i]}")
    i += 1
print("Conclusion: Therefore, Socrates is mortal")

# List comprehension - efficient way to create lists
# Extract first word from each school
first_words = [school.split()[0] for school in schools_of_thought]
print(f"\nFirst words: {first_words}")
```

Functions let us create reusable blocks of code - essential for avoiding repetition and organizing our programs.

```python
# Working with CSV and JSON formats
import json

# Sample data
philosophers_data = """Name,Birth,Death,School
Plato,-428,-348,Idealism
Aristotle,-384,-322,Realism
Kant,1724,1804,Transcendental Idealism
Mill,1806,1873,Utilitarianism"""

# Save and read CSV
with open('philosophers.csv', 'w') as f:
    f.write(philosophers_data)

# Read CSV into list of dictionaries
philosophers = []
with open('philosophers.csv', 'r') as f:
    lines = f.readlines()
    headers = lines[0].strip().split(',')
    for line in lines[1:]:
        values = line.strip().split(',')
        phil = dict(zip(headers, values))
        phil['Birth'] = int(phil['Birth'])
        phil['Death'] = int(phil['Death'])
        philosophers.append(phil)

# Display CSV data
print("From CSV:")
for p in philosophers:
    print(f"- {p['Name']}: {p['Death'] - p['Birth']} years")

# Save as JSON (more flexible format)
with open('philosophers.json', 'w') as f:
    json.dump(philosophers, f, indent=2)

# Read from JSON
with open('philosophers.json', 'r') as f:
    data = json.load(f)
    
print("\nFrom JSON:")
print(f"Total philosophers: {len(data)}")
print(f"First entry: {data[0]['Name']} ({data[0]['School']})")
```

## 6. Final Project: Philosophy Research Assistant (20 min)

Combine everything you've learned to build a comprehensive research tool.

```python
# Philosophy Research Assistant (Concise Version)

class PhilosophyResearchAssistant:
    def __init__(self):
        self.philosophers = []
        self.quotes = []
        self.concepts = set()

    def add_philosopher(self, name, birth, death, school, works):
        """Add a philosopher."""
        self.philosophers.append({
            "name": name,
            "birth": birth,
            "death": death,
            "school": school,
            "works": works,
            "era": self._era(birth)
        })

    def _era(self, year):
        if year < 0:
            return "Ancient"
        elif year < 500:
            return "Late Ancient"
        elif year < 1400:
            return "Medieval"
        elif year < 1800:
            return "Modern"
        else:
            return "Contemporary"

    def add_quote(self, text, author, source="Unknown"):
        """Add a quote."""
        self.quotes.append({
            "text": text.strip(),
            "author": author,
            "source": source
        })
        for word in text.lower().split():
            if len(word) > 6:
                self.concepts.add(word)

    def find_philosophers(self, era):
        """List philosophers from an era."""
        return [p for p in self.philosophers if p["era"] == era]

    def search_quotes(self, keyword):
        """Find quotes containing a keyword."""
        return [q for q in self.quotes if keyword.lower() in q["text"].lower()]

    def summary(self):
        """Quick database summary."""
        return {
            "philosophers": len(self.philosophers),
            "quotes": len(self.quotes),
            "concepts": len(self.concepts)
        }

# Example usage
assistant = PhilosophyResearchAssistant()
assistant.add_philosopher("Plato", -428, -348, "Idealism", ["Republic"])
assistant.add_philosopher("Kant", 1724, 1804, "Idealism", ["Critique of Pure Reason"])
assistant.add_quote("The unexamined life is not worth living", "Socrates")
assistant.add_quote("I think therefore I am", "Descartes")

print("Summary:", assistant.summary())
print("Modern philosophers:", [p["name"] for p in assistant.find_philosophers("Modern")])
print("Quotes with 'life':", [q["text"] for q in assistant.search_quotes("life")])
```
